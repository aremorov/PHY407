```python
import numpy as np
import matplotlib.pyplot as plt
from time import time

N_list = np.arange(2, 250)
t_list1 = []
t_list2 = []

for N in N_list:
    A = np.ones([N,N], float) * 3
    B = np.ones([N,N], float) * 2
    C1 = np.zeros([N,N])
    
    #FIRST METHOD:________________
    
    #save start time
    start=time()
    
    for i in range(N):
        for j in range(N):
            for k in range(N):
                C1[i,j] += A[i,k]*B[k,j]
                
    #save end time            
    end=time()
    
    diff = end-start
    
    t_list1.append(diff)
    
    #SECOND METHOD:_______________
    
    #save start time
    start=time()

    C = np.dot(A,B)
                
    #save end time            
    end=time()
    
    diff = end-start
    
    t_list2.append(diff)
    
    
```

```python
t_list1 = np.array(t_list1)
t_list2 = np.array(t_list2)
```

```python
plt.plot(t_list1)
plt.xlabel("N")
plt.ylabel("Time (s)")
plt.title("Time vs. row length N using Method 1")
plt.savefig("Q3_m1_tN.pdf")
```

```python
plt.plot(N_list**3, t_list1)
plt.xlabel("N^3")
plt.ylabel("Time (s)")
plt.title("Time vs. row length N^3 using Method 1")
plt.savefig("Q3_m1_tN3.pdf")
```

```python
plt.plot(t_list2)
plt.xlabel("N")
plt.ylabel("Time (s)")
plt.title("Time vs. row length N using Method 2")
plt.savefig("Q3_m2_tN.pdf")
```

```python
plt.plot(N_list**3, t_list2)
plt.xlabel("N^3")
plt.ylabel("Time (s)")
plt.title("Time vs. row length N^3 using Method 2")
plt.savefig("Q3_m2_tN3.pdf")
```
